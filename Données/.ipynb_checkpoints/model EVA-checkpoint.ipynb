### Modèle tranche1 : classification multiclasse

```python
dataset = tranche1
```

```python
# séparation des variables prédictives (X) et à prédire (y)
X = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

```python
# création du modèle
automl = AutoML()

# paramètres de automl
automl_settings = {
    "time_budget": 60,  # in seconds (24h = 86400 secondes)
    "metric": 'accuracy',
    "task": 'classification',
    "n_jobs": -1,
    "estimator_list": ['lgbm', 'xgboost'],
    "eval_method": 'cv',
    "n_splits": 5,
    "verbose": 0
    
}

# entrainement de automl
automl.fit(X_train=X, y_train=y, **automl_settings)
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```python
# évaluation du modèle : accuracy
results = cross_val_score(automl.model, X, y, cv=kfold, scoring='accuracy')
results.mean()
```

### Modèle tranche1 : classification binaire

```python
dataset = tranche1
dataset = dataset[dataset['GroupeAppartenance'] != '1.0']
dataset.reset_index(drop=True, inplace=True)
```

```python
# séparation des variables prédictives (X) et à prédire (y)
X = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

```python
# création du modèle
automl2 = AutoML()

# paramètres de automl
automl_settings = {
    "time_budget": 60,  # in seconds (24h = 86400 secondes)
    "metric": 'accuracy',
    "task": 'classification',
    "n_jobs": -1,
    "estimator_list": ['lgbm', 'xgboost'],
    "eval_method": 'cv',
    "n_splits": 5,
    "verbose": 0
    
}

# entrainement de automl
automl2.fit(X_train=X, y_train=y, **automl_settings)
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```python
# évaluation du modèle : accuracy
results = cross_val_score(automl2.model, X, y, cv=kfold, scoring='accuracy')
results.mean()
```
