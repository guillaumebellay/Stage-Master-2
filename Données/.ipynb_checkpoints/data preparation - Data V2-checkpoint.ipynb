```python
import pandas as pd
import numpy as np
```

```python
# chargement des données
data = pd.read_csv('Data-V1.txt', index_col=0, dtype='Int64')
data.head(10)
```

```python
data.iloc[:,503:]
```

```python
# percentage of missing values
pct_nan = pd.Series(data.isna().sum().sort_values(ascending=False) / 586.0)
pct_nan
```

```python
for pct in np.arange(0,1,0.1):
    print('pct {} = {}'.format(pct,sum(pct_nan >= pct)))
```

```python
# suppression des variables avec plus de 90% de valeurs nulles ?
```

```python
# imputation avec l'algorithme des K-moyennes
from sklearn.impute import KNNImputer
```

```python
# les voisins les plus proches ont une plus grande influence que les voisins les plus éloignés.
# on prend les 5 voisins les plus proches. Paramètre à faire varier.
knn = KNNImputer(n_neighbors=5, weights='distance')
```

```python
data_trans = pd.DataFrame(np.round(knn.fit_transform(data)), columns=data.columns, dtype='Int64')
data_trans
```
